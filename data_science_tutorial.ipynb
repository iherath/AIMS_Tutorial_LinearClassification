# Data Exploration 
- With any data science/ ML project it is always best to get an understanding of the distribution of data

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime


# Set the style for the plots
plt.style.use('ggplot')
sns.set(style="whitegrid")

# Define the data directory
data_dir = '10k_synthea_covid19_csv'

# Check if the data directory exists
if not os.path.exists(data_dir):
    print(f"Error: Data directory '{data_dir}' not found.")
    print("Please ensure you have downloaded the Synthea COVID-19 dataset and placed it in the correct location.")
    print("You can download the dataset from: https://synthea.mitre.org/downloads")
else:
    try:
        # Get all CSV files in the directory
        csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]
        
        # Create paths for all CSV files
        file_paths = {}
        for file in csv_files:
            file_name = os.path.splitext(file)[0]  # Remove .csv extension
            file_paths[file_name] = os.path.join(data_dir, file)
        
        # Exclude payers, payer_transitions, providers
        excluded_files = ['payers', 'payer_transitions', 'providers']
        for excluded in excluded_files:
            if excluded in file_paths:
                del file_paths[excluded]
        
        # For backward compatibility, ensure we have variables for the main files
        patients_path = file_paths.get('patients', '')
        conditions_path = file_paths.get('conditions', '')
        observations_path = file_paths.get('observations', '')
        medications_path = file_paths.get('medications', '')
        careplans_path = file_paths.get('careplans', '')
        encounters_path = file_paths.get('encounters', '')
        procedures_path = file_paths.get('procedures', '')
        
        # Check if individual files exist before loading
        files_to_check = file_paths
        
        missing_files = [name for name, path in files_to_check.items() if not os.path.exists(path)]
        
        if missing_files:
            print(f"Error: The following data files are missing: {', '.join(missing_files)}")
            print(f"Please ensure all required CSV files are in the '{data_dir}' directory.")
        else:
            # Load all datasets
            datasets = {}
            for name, path in file_paths.items():
                datasets[name] = pd.read_csv(path)
            
            # Assign main datasets to variables for backward compatibility
            patients = datasets.get('patients', pd.DataFrame())
            conditions = datasets.get('conditions', pd.DataFrame())
            observations = datasets.get('observations', pd.DataFrame())
            medications = datasets.get('medications', pd.DataFrame())
            care_plans = datasets.get('careplans', pd.DataFrame())
            encounters = datasets.get('encounters', pd.DataFrame())
            procedures = datasets.get('procedures', pd.DataFrame())
            
            # Display basic information about all datasets
            print("Datasets loaded successfully!")
            for name, df in datasets.items():
                print(f"{name.capitalize()}: {df.shape[0]} records with {df.shape[1]} columns")
            
            # Display the first few rows of each dataset
            for name, df in datasets.items():
                print(f"\n{name.capitalize()} sample:")
                display(df.head())
            
            # Basic demographic analysis
            print("\nDemographic Analysis:")
            
            # Age distribution - fixed to handle the timedelta conversion error
            if 'BIRTHDATE' in patients.columns:
                patients['BIRTHDATE'] = pd.to_datetime(patients['BIRTHDATE'])
                current_date = datetime.now()
                # Calculate age in years using datetime objects instead of timedelta
                patients['AGE'] = patients['BIRTHDATE'].apply(lambda x: current_date.year - x.year - ((current_date.month, current_date.day) < (x.month, x.day)))
                
                plt.figure(figsize=(10, 6))
                sns.histplot(patients['AGE'], bins=20, kde=True)
                plt.title('Age Distribution')
                plt.xlabel('Age')
                plt.ylabel('Count')
                plt.show()
                
                # Age statistics
                print(f"Age statistics:\n{patients['AGE'].describe()}")
            
            # Gender distribution
            if 'GENDER' in patients.columns:
                plt.figure(figsize=(8, 6))
                gender_counts = patients['GENDER'].value_counts()
                gender_counts.plot(kind='bar')
                plt.title('Gender Distribution')
                plt.xlabel('Gender')
                plt.ylabel('Count')
                plt.xticks(rotation=0)
                plt.show()
                
                # Gender percentage
                print(f"Gender distribution:\n{gender_counts / len(patients) * 100}")
            
            # Condition analysis
            if 'DESCRIPTION' in conditions.columns:
                plt.figure(figsize=(12, 8))
                top_conditions = conditions['DESCRIPTION'].value_counts().head(10)
                top_conditions.plot(kind='barh')
                plt.title('Top 10 Conditions')
                plt.xlabel('Count')
                plt.ylabel('Condition')
                plt.tight_layout()
                plt.show()
                
                print(f"\nTop 10 conditions:\n{top_conditions}")
    
    except FileNotFoundError as e:
        print(f"Error loading data: {e}")
        print("Please check the file paths and ensure the data files are available.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

# Creating Basic Models for COVID Status Prediction

Train-Test Splits and Validation Sets

Train-test splits divide our data into separate sets: a training set to build the model and a test set to evaluate its performance on unseen data. This helps assess how well our model generalizes. Typically, we use 70-80% for training and 20-30% for testing.

Validation sets provide an additional layer of evaluation during model development. They help tune hyperparameters and prevent overfitting without contaminating the test set. Common approaches include k-fold cross-validation, where data is split into k subsets, with each subset serving as a validation set while the rest are used for training.


## Decision Tree
Decision trees are intuitive machine learning models that make predictions by following a series of yes/no questions. They work similar to a flowchart, where each internal node represents a "test" on a feature (e.g., "Is age > 65?"), each branch represents the outcome of that test, and each leaf node represents a prediction (e.g., "Survivor" or "Non-survivor").

Advantages of decision trees:
 - Easy to understand and interpret (even for non-technical people)
 - Can handle both numerical and categorical data
 - Require minimal data preparation
 - Can be visualized to clearly show the decision-making process

Limitations of decision trees:
 - Prone to overfitting, especially with deep trees
 - Can be unstable - small variations in data might result in a completely different tree
 - May create biased trees if some classes dominate
 - Not ideal for capturing complex relationships in the data

To overcome these limitations, ensemble methods like bagging and gradient boosting have been developed:

**Bagging (Bootstrap Aggregating)**: Creates multiple decision trees by sampling with replacement from the training data and averages their predictions. Random Forest is a popular bagging algorithm that also randomly selects features at each split.

**Gradient Boosting**: Builds trees sequentially, with each tree correcting the errors of the previous ones. XGBoost (Extreme Gradient Boosting) is a highly efficient implementation that offers improved speed and performance, making it one of the most popular algorithms for structured data in machine learning competitions.

In our COVID analysis, a decision tree will help us identify which lab values and patient characteristics are most important for predicting survival outcomes, and the specific thresholds that separate survivors from non-survivors.
For more information and theory please refer to [this website](https://www.cs.cornell.edu/courses/cs4780/2018fa/lectures/lecturenote17.html)


```python
# Import necessary libraries for decision tree modeling
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib.colors import ListedColormap

# Select features and target variable
X = patient_labs_df.drop(['PATIENT', 'survivor'], axis=1)
y = patient_labs_df['survivor']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a decision tree classifier with controlled complexity
dt_classifier = DecisionTreeClassifier(max_depth=3, min_samples_split=10, 
                                      min_samples_leaf=5, random_state=42)

# Train the model
dt_classifier.fit(X_train, y_train)

# Make predictions
y_pred = dt_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Display confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Non-survivor', 'Survivor'],
            yticklabels=['Non-survivor', 'Survivor'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Visualize the decision tree with better formatting
plt.figure(figsize=(15, 10))
tree.plot_tree(dt_classifier, 
               feature_names=X.columns, 
               class_names=['Non-survivor', 'Survivor'],
               filled=True, 
               rounded=True,
               precision=2,
               proportion=False,
               fontsize=10)
plt.title("Decision Tree for COVID Survival Prediction")
plt.tight_layout()
plt.show()

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': dt_classifier.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
plt.title('Feature Importance for Survival Prediction')
plt.tight_layout()
plt.show()

# Plot decision boundary with known ferritin threshold of 727.35
# Find the second most important feature for visualization
top_features = feature_importance['Feature'].head(2).tolist()
# Ensure 'Ferritin' is the first feature
if 'Ferritin' not in top_features:
    top_features = ['Ferritin'] + [f for f in top_features if f != 'Ferritin'][:1]
elif top_features[0] != 'Ferritin':
    top_features = ['Ferritin', top_features[0]]

X_important = X[top_features]
X_train_important = X_train[top_features]
X_test_important = X_test[top_features]

# Create a mesh grid to plot the decision boundary
h = 0.02  # step size in the mesh
x_min, x_max = X_important.iloc[:, 0].min() - 1, X_important.iloc[:, 0].max() + 1
y_min, y_max = X_important.iloc[:, 1].min() - 1, X_important.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Create a custom decision boundary function using the known ferritin threshold
def custom_predict(X_subset):
    predictions = np.zeros(X_subset.shape[0])
    # If Ferritin > 727.35, predict non-survivor (0), else survivor (1)
    predictions[X_subset[:, 0] > 727.35] = 0
    predictions[X_subset[:, 0] <= 727.35] = 1
    return predictions

# Predict class for each point in the mesh using the known threshold
Z = custom_predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])

# Plot the decision boundary
fig, ax = plt.subplots(figsize=(10, 8))
contour = ax.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
scatter = ax.scatter(X_test_important.iloc[:, 0], X_test_important.iloc[:, 1], c=y_test, 
           cmap=cmap_bold, edgecolor='k', s=20)
ax.set_xlabel(top_features[0])
ax.set_ylabel(top_features[1])
ax.set_title(f'Decision Boundary with Ferritin Threshold = 727.35')

# Add a vertical line at the ferritin threshold
ax.axvline(x=727.35, color='black', linestyle='--', label='Ferritin Threshold = 727.35')
ax.legend()

ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())

# Create a ScalarMappable with the colormap
sm = plt.cm.ScalarMappable(cmap=cmap_bold)
sm.set_array([0, 1])
fig.colorbar(sm, ax=ax, label='Survived', ticks=[0.25, 0.75], boundaries=[0, 0.5, 1])
plt.tight_layout()
plt.show()

print(f"Number of samples used for training: {len(X_train)}")
print(f"Number of features used: {X.shape[1]}")
print(f"Known decision boundary for Ferritin: 727.35")

```

This maybe an issue with our sample. However, note the decision boundary plot. This is a good way of visualizing our model:
- The background color corresponds to the model prediction (so any points landing in the red region will be predicted as "does not survive"). The transition from green to red regions marks the threshold calculated by the decision tree as is reflected in the plot of the actual tree.
- The color of the points themselves represent the true value of the outcome

## Support Vector Machines (SVMs)
# <img src="./svm.png" width="400">


Support Vector Machines are powerful supervised learning models used for classification and regression tasks. The key idea behind SVMs is to find the optimal hyperplane that separates different classes with the maximum margin.

### How SVMs work:

1. **Find the optimal hyperplane**: SVM tries to find a decision boundary that maximizes the distance between the closest points (support vectors) of different classes.

2. **Margin maximization**: The algorithm maximizes the margin between classes, which helps improve generalization to new data.

3. **Kernel trick**: SVMs can handle non-linear data by using kernel functions (like RBF, polynomial) that transform the data into higher dimensions where it becomes linearly separable.

## Advantages of SVMs:

- Effective in high-dimensional spaces
- Memory efficient as they use only a subset of training points (support vectors)
- Versatile through different kernel functions
- Robust against overfitting, especially in high-dimensional spaces

## Limitations:

- Can be computationally intensive for large datasets
- Requires careful tuning of hyperparameters
- Less interpretable than decision trees

For a more detailed mathematical explanation of Support Vector Machines, check out [this lecture note from Cornell University](https://www.cs.cornell.edu/courses/cs4780/2018fa/lectures/lecturenote09.html).

## Logistic Regression
# ![Logistic Regression](logreg.png)

Logistic Regression is a fundamental statistical model used for binary classification problems. Despite its name, it's a classification algorithm rather than a regression algorithm, predicting the probability that an instance belongs to a particular class.

### How Logistic Regression works:

1. **Linear combination**: The algorithm creates a linear combination of the input features.

2. **Sigmoid transformation**: It applies the sigmoid function to transform the linear combination into a probability value between 0 and 1.

3. **Decision boundary**: A threshold (typically 0.5) is used to convert probabilities into binary class predictions.

### Advantages of Logistic Regression:

- Simple to implement and interpret
- Efficient training process
- Provides probability scores rather than just classifications
- Less prone to overfitting in low-dimensional datasets
- Works well for linearly separable data

### Limitations:

- Assumes a linear relationship between features and log-odds
- May underperform with complex non-linear relationships
- Cannot solve non-linear problems without feature engineering
- Sensitive to imbalanced datasets

For a more detailed explanation of Logistic Regression, you can refer to [this course note](https://www.cs.cornell.edu/courses/cs4780/2024sp/lectures/lecturenote06.html).

## Perceptron
# <img src="./perceptron.png" width="400">
 
The Perceptron is one of the earliest and simplest types of artificial neural networks (invented by Frank Rosenblatt in 1958 at Cornell ;)). It's a binary classifier that makes predictions based on a linear predictor function combining a set of weights with the feature vector.
 
### How the Perceptron works:

1. **Input processing**: The algorithm takes a set of input features and multiplies each by a weight.

2. **Summation**: It sums these weighted inputs along with a bias term.

3. **Activation function**: The sum is passed through a step function that outputs 1 if the sum is positive and 0 (or -1) otherwise.

4. **Learning**: When the perceptron makes an error, it updates the weights to correct the prediction.

### Advantages of the Perceptron:

- Conceptually simple and easy to implement
- Computationally efficient
- Guaranteed to converge if the data is linearly separable
- Forms the foundation for more complex neural networks
- Online learning capability (can update with new data)

### Limitations:

- Can only learn linearly separable patterns
- Cannot solve XOR and other non-linear problems
- Sensitive to feature scaling
- May never converge if data is not linearly separable
- Limited expressiveness compared to modern neural networks

For a more detailed mathematical explanation of the Perceptron algorithm, check out [this lecture note from Cornell University](https://www.cs.cornell.edu/courses/cs4780/2024sp/lectures/lecturenote03.html).

# Comparing Models with ROC curves

## Understanding ROC Curves and AUC

ROC (Receiver Operating Characteristic) curves are a powerful tool for evaluating and comparing classification models. They plot the True Positive Rate (TPR, or sensitivity) against the False Positive Rate (FPR, or 1-specificity) at various classification thresholds.

### Key components:
- **True Positive Rate (TPR)**: The proportion of actual positives correctly identified (TP/(TP+FN))
- **False Positive Rate (FPR)**: The proportion of actual negatives incorrectly classified as positive (FP/(FP+TN))

### AUC (Area Under the Curve):
- AUC measures the entire two-dimensional area under the ROC curve
- AUC ranges from 0 to 1, where:
  - 0.5 suggests no discrimination ability (equivalent to random guessing)
   - Greater than 0.7 is considered acceptable
   - Greater than 0.8 is considered excellent
   - Greater than 0.9 is considered outstanding

### How to interpret ROC curves:
- The closer the curve follows the left-top border of the ROC space, the more accurate the model
- The closer the curve comes to the 45-degree diagonal, the less useful the model
- When comparing models, the one with the higher AUC generally performs better overall

ROC curves are particularly useful when working with imbalanced datasets or when different misclassification costs need to be considered.

```python
# Plot ROC curves for all models
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(10, 8))

# SVM model
y_score_svm = svm_classifier.decision_function(X_test)
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_score_svm)
roc_auc_svm = auc(fpr_svm, tpr_svm)
plt.plot(fpr_svm, tpr_svm, label=f'SVM (AUC = {roc_auc_svm:.2f})', linewidth=2)

# Logistic Regression model
y_score_lr = log_reg.predict_proba(X_test)[:, 1]
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_score_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.2f})', linewidth=2)

# Decision Tree model
y_score_dt = dt_classifier.predict_proba(X_test)[:, 1]
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_score_dt)
roc_auc_dt = auc(fpr_dt, tpr_dt)
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {roc_auc_dt:.2f})', linewidth=2)

# Perceptron model - using the full model
# We need to ensure we're using the same features as during training
try:
    # For the regular perceptron (not the simple one)
    y_score_perceptron = perceptron.decision_function(X_test)
    fpr_perceptron, tpr_perceptron, _ = roc_curve(y_test, y_score_perceptron)
    roc_auc_perceptron = auc(fpr_perceptron, tpr_perceptron)
    plt.plot(fpr_perceptron, tpr_perceptron, label=f'Perceptron (AUC = {roc_auc_perceptron:.2f})', linewidth=2)
except Exception as e:
    print(f"Error with regular perceptron: {e}")
    roc_auc_perceptron = 0

# Add the simplified perceptron model
# Make sure to use X_test_important which has the same features used during training
try:
    y_score_perceptron_simple = perceptron_simple.decision_function(X_test_important)
    fpr_perceptron_simple, tpr_perceptron_simple, _ = roc_curve(y_test, y_score_perceptron_simple)
    roc_auc_perceptron_simple = auc(fpr_perceptron_simple, tpr_perceptron_simple)
    plt.plot(fpr_perceptron_simple, tpr_perceptron_simple, label=f'Simple Perceptron (AUC = {roc_auc_perceptron_simple:.2f})', linewidth=2)
except Exception as e:
    print(f"Error with simple perceptron decision_function: {e}")
    # Fallback to predict if decision_function is not available
    try:
        y_score_perceptron_simple = perceptron_simple.predict(X_test_important)
        fpr_perceptron_simple, tpr_perceptron_simple, _ = roc_curve(y_test, y_score_perceptron_simple)
        roc_auc_perceptron_simple = auc(fpr_perceptron_simple, tpr_perceptron_simple)
        plt.plot(fpr_perceptron_simple, tpr_perceptron_simple, label=f'Simple Perceptron (AUC = {roc_auc_perceptron_simple:.2f})', linewidth=2)
    except Exception as e:
        print(f"Error with simple perceptron predict: {e}")
        roc_auc_perceptron_simple = 0

# Add diagonal line representing random classifier
plt.plot([0, 1], [0, 1], 'k--', label='Random (AUC = 0.50)')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different Models')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print summary of model performance
print("Model Performance Summary (AUC-ROC):")
print(f"SVM: {roc_auc_svm:.4f}")
print(f"Logistic Regression: {roc_auc_lr:.4f}")
print(f"Decision Tree: {roc_auc_dt:.4f}")
if roc_auc_perceptron > 0:
    print(f"Perceptron: {roc_auc_perceptron:.4f}")
if roc_auc_perceptron_simple > 0:
    print(f"Simple Perceptron (only using lymphocytes and age): {roc_auc_perceptron_simple:.4f}")
```

- These models are all performing well, but we can see that the Decision tree is performing the best. We got luckly because the data seems pretty linearly separable making all these models work pretty well. 
- For data that is not linearly separable, we would need to consider different approaches:
  1. [Kernel methods](https://www.cs.cornell.edu/courses/cs5780/2025sp/lectures/lecturenote13.html): SVM with non-linear kernels (RBF, polynomial) can transform data into higher dimensions where it becomes separable.
  2. [Neural networks](https://www.cs.cornell.edu/courses/cs4780/2022sp/notes/LectureNotes23.html): Multi-layer perceptrons can learn complex non-linear decision boundaries
  3. Ensemble methods: [Random Forests](https://www.cs.cornell.edu/courses/cs4780/2024sp/lectures/lecturenote18.html) or [Gradient Boosting](https://www.cs.cornell.edu/courses/cs4780/2024sp/lectures/lecturenote19.html) can capture non-linear patterns through multiple decision trees
  4. Feature engineering: Creating new features through transformations or interactions can help make non-linear problems more separable
  5. Dimensionality reduction: [PCA](https://www.cs.cornell.edu/courses/cs4780/2024sp/lectures/UnsupervisedLearning.html#pca), t-SNE or UMAP can help visualize and potentially simplify complex non-linear relationships
